# Метрики для подсчёта качества


Так как часть метрик способны дать лишь отдельные слова, а не словосочетания мы будем считать точность, полноту, F-меру  по отдельным словам, а не по выражениям.

```python
def evaluate(ass_keys, true_keys, digit):
    TP = 0
    FN= 0
    FP = 0
    for word in ass_keys:
        if word in true_keys:
            TP +=1
        else:
            FN += 1
    for word in true_keys:
        if word not  in ass_keys:
            FP +=1
    if (TP == 0) and (FP == 0):
        prec= 0
    else:
        prec = TP/(TP+FP)
    if (TP == 0) and (FN == 0):
        rec = 0
    else:
        rec = TP/(TP+FN)
    if (prec == 0) and (rec == 0):
        F= 0
    else:
        F = 2 * ((prec*rec)/(prec+rec))
    print('precision text', digit, ': ', prec)
    print('recall text', digit, ': ', rec)
    print('F-metrics text',digit,':', F)
    print('____________________')
    return
```

```python
from nltk.parse import DependencyGraph
```

```python
_FILTER_RELS = ['punct', 'conj', 'parataxis']
def get_subtree(nodes, node):
    if not nodes[node]['deps']:
        return [node]
    else:
        return [node] + [get_subtree(nodes, dep) for rel in nodes[node]['deps'] 
                         if rel not in _FILTER_RELS
                         for dep in nodes[node]['deps'][rel]]  
```

```python
def nominals(name):
    trees = []

    with open(name, encoding = 'utf-8') as f:
        parsed_sents = f.read().split('\n\n')

    for sent in parsed_sents:
        tree = [line for line in sent.split('\n') if line and line[0] != '#']
        trees.append('\n'.join(tree))
        
    result = []
    for t in trees:
        g = DependencyGraph(t, top_relation_label='root')
        np_list = []
        for n in g.nodes:
            if g.nodes[n]['ctag'] == 'NOUN':
                np = list(sorted(flatten(get_subtree(g.nodes, n))))
                np_list.append(
                ' '.join([g.nodes[i]['word'] for i in np])
                )
        result.append(np_list)
    return result
```

# BM25

Оно же tf_idf на максималках

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from math import log

vectorizer1 = CountVectorizer(binary = True)
vectorizertf = TfidfVectorizer(use_idf=False)

lenghths = [len(text_1), len(text_2), len(text_3), len(text_4)]
avgdl = np.mean(lenghths) 
k = 2
b = 0.75
```

```python
corpus_matrix1 = vectorizer1.fit_transform([text_1.lower(), text_2.lower(), text_3.lower(), text_4.lower()])
voc = vectorizer1.get_feature_names()
N = len(corpus_matrix1.getnnz(axis = 1))
n = corpus_matrix1.sum(axis=0)
n = n.tolist()[0]
```

```python
idf = []
for i in range(len(voc)):
    idf_p = log((N-n[i]+0.5) / (n[i]+0.5))
    if idf_p < 0:
            idf_p = 0
    idf.append(idf_p)
```

```python
tf = vectorizertf.fit_transform([text_1, text_2, text_3, text_4])
tf = tf.toarray()

additional = []
for i  in lenghths:
    res = k * (1-b+b*(i/avgdl))
    additional.append(res)
```

```python
k1 = k + 1
for i in range(N):
    for j in range(len(voc)):
        tf[i][j] = (tf[i][j]*k1)/(tf[i][j] + additional[i]) * idf[j]
```

Таким образом, в переменной tf мы получаем матрицу со значениями bm25
